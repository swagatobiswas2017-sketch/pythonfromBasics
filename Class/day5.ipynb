```python
import time as t

class Pomodoro:
    def __init__(self, dur1: int, dur2: int):
        st: int = dur1*60
        bre: int = dur2*60
        self.s = st
        self.b = bre
    def start(self):
        print("Study Time Starts: ")
        remaining = self.s
        while remaining > 0:
            print(f"Time remaining : {remaining}")
            t.sleep(1)
            remaining -= 1
        print("Study Time Ends:")
            
    def leasure(self):
        print("Break Time Starts: ")
        remaining = self.b
        while remaining > 0:
            print(f"Time remaining : {remaining}")
            t.sleep(1)
            remaining -= 1
        print("Break Time Ends: ")
    
    


if __name__ == "__main__":
   pomo = Pomodoro(1,0.5)
   pomo.start()
   pomo.leasure()
```
