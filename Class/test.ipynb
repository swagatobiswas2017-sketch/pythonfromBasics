```python
class Vehicle:
    def __init__(self, wheels, speedometer, engine, fuel):
        self.wheels = wheels
        self.speedometer = speedometer
        self.engine = engine
        self.fuel = fuel

    def start(self):
        self.engine.start()
        self.fuel.start_pump()

    def stop(self):
        self.engine.stop()
        self.fuel.stop_pump()

    def __str__(self):
        return f"Vehicle with {self.wheels} wheels and {self.speedometer} speedometer"

# Inheritance

class Car(Vehicle):
    def __init__(self, wheels, speedometer, engine, fuel, windows):
        super().__init__(wheels, speedometer, engine, fuel)
        self.windows = windows
        
    def __str__(self):
        return super().__str__() + f" and {self.windows} windows"

class Bike(Vehicle):
    def __init__(self, wheels, speedometer, engine, fuel, pedals):
        super().__init__(wheels, speedometer, engine, fuel)
        self.pedals = pedals
        
    def __str__(self):
        return super().__str__() + f" and {self.pedals} pedals"
```
