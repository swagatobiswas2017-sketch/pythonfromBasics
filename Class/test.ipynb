```python
class Vehicle:
    def __init__(self, wheels, speedometer, engine, fuel):
        self.wheels = wheels
        self.speedometer = speedometer
        self.engine = engine
        self.fuel = fuel

    def start(self):
        self.engine.start()
        self.fuel.start_pump()

    def stop(self):
        self.engine.stop()
        self.fuel.stop_pump()

    def __str__(self):
        return f"Vehicle with {self.wheels} wheels and {self.speedometer} speedometer"

# Inheritance

class Car(Vehicle):
    def __init__(self, wheels, speedometer, engine, fuel, windows):
        super().__init__(wheels, speedometer, engine, fuel)
        self.windows = windows
        
    def __str__(self):
        return super().__str__() + f" and {self.windows} windows"

class Bike(Vehicle):
    def __init__(self, wheels, speedometer, engine, fuel, pedals):
        super().__init__(wheels, speedometer, engine, fuel)
        self.pedals = pedals
        
    def __str__(self):
        return super().__str__() + f" and {self.pedals} pedals"
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Create grid
alpha = np.linspace(-2, 3, 500)
beta = np.linspace(-2, 3, 500)
A, B = np.meshgrid(alpha, beta)

# Inequality condition
Z = A + B - A * B

# Create figure
plt.figure(figsize=(8, 6))

# Shade region satisfying inequality
plt.contourf(A, B, Z >= 0.5, levels=[0.5, 1], colors=['lightgrey'])

# Plot boundary curve where equality holds
# Solve for beta in terms of alpha:
# alpha + beta - alpha*beta = 0.5
# beta(1 - alpha) = 0.5 - alpha
# beta = (0.5 - alpha)/(1 - alpha)
alpha_curve = np.linspace(-2, 3, 1000)
epsilon = 1e-6  # avoid division by zero at alpha=1
alpha_curve = alpha_curve[np.abs(alpha_curve - 1) > epsilon]
beta_curve = (0.5 - alpha_curve) / (1 - alpha_curve)

plt.plot(alpha_curve, beta_curve, color='black')

# Labels
plt.xlabel(r'$\alpha_1$', fontsize=12)
plt.ylabel(r'$\beta_1$', fontsize=12)

# Legend
legend_patch = Patch(facecolor='lightgrey', label=r'$\alpha_1 + \beta_1 - \alpha_1 \beta_1 \geq 0.5$')
plt.legend(handles=[legend_patch])

plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.title("Feasible Region of Inequality")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Create grid
alpha = np.linspace(-2, 3, 500)
beta = np.linspace(-2, 3, 500)
A, B = np.meshgrid(alpha, beta)

# Inequality condition
Z = A + B - A * B

# Create figure
fig, ax = plt.subplots(figsize=(8, 6))

# Shade region satisfying inequality
ax.contourf(A, B, Z >= 0.5, levels=[0.5, 1], colors=['lightgrey'])

# Boundary curve
alpha_curve = np.linspace(-2, 3, 1000)
epsilon = 1e-6
alpha_curve = alpha_curve[np.abs(alpha_curve - 1) > epsilon]
beta_curve = (0.5 - alpha_curve) / (1 - alpha_curve)

ax.plot(alpha_curve, beta_curve, color='black')

# ---- SHIFT AXES TO ORIGIN ----
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Axis labels
ax.set_xlabel(r'$\alpha_1$', fontsize=12)
ax.set_ylabel(r'$\beta_1$', fontsize=12)

# Legend
legend_patch = Patch(facecolor='lightgrey',
                     label=r'$\alpha_1 + \beta_1 - \alpha_1 \beta_1 \geq 0.5$')
ax.legend(handles=[legend_patch], loc='upper right')

ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)

ax.grid(True)
plt.title("Feasible Region of Inequality")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# XOR dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y = np.array([0, 1, 1, 0], dtype=float)

# Center and convert to complex
z = (X[:,0] - 0.5) + 1j*(X[:,1] - 0.5)

# Initialize complex weight
np.random.seed(0)
w = np.random.randn() + 1j*np.random.randn()

lr = 0.5
epochs = 3000

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

for _ in range(epochs):

    # Forward
    u = w * z
    f = np.real(u**2)   # key nonlinear geometric operation
    y_hat = sigmoid(f)

    error = y_hat - y

    # derivative of Re(u^2) wrt w
    grad = np.sum(error * 2 * z * u)

    w -= lr * grad / len(z)

# Final prediction
u = w * z
f = np.real(u**2)
y_hat = sigmoid(f)
predictions = (y_hat > 0.5).astype(int)

print("Predictions:", predictions)
print("Targets:", y)

plt.figure()
plt.scatter(f, y)
plt.xlabel("Re((wz)^2)")
plt.ylabel("Class")
plt.title("XOR via Angular Complex Squaring")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# XOR dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y = np.array([0, 1, 1, 0], dtype=float)
y_pm = 2*y - 1

z = X[:,0] + 1j * X[:,1]

np.random.seed(0)
w = np.exp(1j*np.random.rand())
b = np.random.randn()*0.1 + 1j*np.random.randn()*0.1

lr = 0.01
epochs = 40000

for _ in range(epochs):

    u = w * z + b
    
    # Remove magnitude
    u_norm = u / (np.abs(u) + 1e-8)
    
    f = np.real(u_norm**2)

    margins = y_pm * f
    active = margins < 1

    grad = np.zeros_like(w, dtype=complex)
    grad_b = 0+0j
    
    for i in range(len(z)):
        if active[i]:
            grad += -2*y_pm[i]*u_norm[i]*z[i]
            grad_b += -2*y_pm[i]*u_norm[i]

    w -= lr * grad / len(z)
    b -= lr * grad_b / len(z)

    # keep w unit magnitude
    w /= np.abs(w)

# Final evaluation
u = w*z + b
u_norm = u/(np.abs(u)+1e-8)
f = np.real(u_norm**2)
pred = (f > 0).astype(int)

print("Weight:", w)
print("Bias:", b)
print("Predictions:", pred)
print("Targets:", y)
print("Accuracy:", np.mean(pred == y))

plt.figure()
plt.scatter(f, y)
plt.xlabel("Re(normalized(u)^2)")
plt.ylabel("Class")
plt.title("Pure Angular XOR Neuron")
plt.show()
```

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y = np.array([0,1,1,0])
y_pm = 2*y - 1

# Center explicitly
z = (X[:,0] - 0.5) + 1j*(X[:,1] - 0.5)

# Only learn rotation angle
theta = 0.3   # small asymmetry
lr = 0.1

for _ in range(20000):

    w = np.exp(1j*theta)
    u = w*z
    f = np.real(u**2)

    margins = y_pm * f
    grad = 0

    for i in range(len(z)):
        if margins[i] < 1:
            # derivative wrt theta
            du_dtheta = 1j*w*z[i]
            df_dtheta = 2*np.real(u[i]*du_dtheta)
            grad += -y_pm[i]*df_dtheta

    theta -= lr * grad / len(z)

# Final
w = np.exp(1j*theta)
f = np.real((w*z)**2)
pred = (f > 0).astype(int)

print("Predictions:", pred)
print("Targets:", y)
print("Accuracy:", np.mean(pred == y))
```
