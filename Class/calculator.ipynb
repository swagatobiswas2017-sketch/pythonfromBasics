```python
class Calculator:
    def __init__(self, n1 : int, n2 : int):
        self.a = n1
        self.b = n2
    def norm_calc(self):
        c = self.a
        d = self.b
        sum = c + d
        sub = c - d
        multi = c * d
        div = c / d
        return c,d,sum,sub,multi,div
no1 = int(input("Enter 1st no. :\n"))
no2 = int(input("Enter 2nd no. :\n"))  
calc = Calculator(n1 = no1, n2 = no2)
num1,num2,ad,mi,mul,by = calc.norm_calc()
print(num1," + ",num2," = ",ad)
print(num1," - ",num2," = ",mi)
print(num1," x ",num2," = ",mul)
print(num1," \ ",num2, " = ",by)
```
